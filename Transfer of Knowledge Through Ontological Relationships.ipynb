# Example Graph

This notebook uses a small example related to [statins](http://www.ebi.ac.uk/chebi/searchId.do?chebiId=CHEBI%3A87631) and their target, [HMG-CoA_reductase](https://en.wikipedia.org/wiki/HMG-CoA_reductase). While it is not exhaustively importing relationships using [Bio2BEL ExPASy](http://bio2bel-expasy.readthedocs.io) and [Bio2BEL ChEBI](http://bio2bel-chebi.readthedocs.io), it still provides an atomic example on how in- and out-edges can be inferred through hierarchies.

```python
to_jupyter(statin_graph)
```

## Propogation on Chemical Hierarchy

The first transformation made is to assert that all things true for EC 1.1.1.34/EC 1.1.88 inhibitors are also true for their descendants. There are two relationships showing decrease in the activities of the corresponding enzyme nodes that will be asserted for its two children, statin and mevinolinic acid, as well as recursively for their descendants.

```python
infer_child_relations(statin_graph, hmgcr_inhibitor)
to_jupyter(statin_graph)
```

## Propogation on Protein Hierarchy

While EC 1.1.1.34 has several proteins across many species annotated in [ExPASy](https://enzyme.expasy.org/EC/1.1.1.34), [HMGCR](https://www.genenames.org/cgi-bin/gene_symbol_report?hgnc_id=HGNC:5006) is normalized to its HGNC nomenclature and used as the only example in this graph. Since HMGCR is an EC 1.1.1.34, all of the relationships with this enzyme node are transferred below.

```python
infer_child_relations(statin_graph, ec_11134)
to_jupyter(statin_graph)
```
